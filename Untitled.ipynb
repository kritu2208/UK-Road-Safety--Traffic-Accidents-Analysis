```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
selected_columns = ["1st_Road_Class","1st_Road_Number","Accident_Severity","Date","Did_Police_Officer_Attend_Scene_of_Accident","Latitude",
                    "Longitude","Number_of_Casualties","Number_of_Vehicles","Pedestrian_Crossing-Human_Control",
                    "Pedestrian_Crossing-Physical_Facilities","Police_Force","Road_Surface_Conditions","Road_Type",
                    "Speed_limit","Time","Weather_Conditions"]
df = pd.read_csv("Accident_Information.csv",usecols=selected_columns)
```

# Data Preprocessing

```python
# Check missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Fill missing values in column with fewer missing values
missing_cols = ['1st_Road_Number', 'Did_Police_Officer_Attend_Scene_of_Accident', 'Latitude', 'Longitude',
               'Pedestrian_Crossing-Human_Control' , 'Pedestrian_Crossing-Physical_Facilities','Speed_limit']

for cols in missing_cols:
    df[cols].fillna(df[cols].median(), inplace=True)

df.isna().sum()
```

```python
# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract useful time features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour 
```

```python
# Convert categorical variables to numerical
df['Accident_Severity'] = df['Accident_Severity'].map({'Slight': 0, 'Serious': 1, 'Fatal': 2})
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert int64 → int32
int_cols = df.select_dtypes(include=['int64']).columns
df[int_cols] = df[int_cols].astype('int32')

# Convert float64 → float32
float_cols = df.select_dtypes(include=['float64']).columns
df[float_cols] = df[float_cols].astype('float32')


df.info(memory_usage='deep')
```

```python
print(df.isnull().sum())
```

```python
df['Hour'] = df['Hour'].fillna(0).astype('int32')
```

```python
df.reset_index(inplace=True)
df.rename(columns = {"index" : "Accident_Index"}, inplace=True)
```

```python
severity_counts=df['Accident_Severity'].value_counts()
severity_counts
```

# EDA(Exploratory Data Analysis)

Visualizing target variable

```python
colors = ['#4682B4', '#5F9EA0', '#9370DB']

plt.figure(figsize=(10, 6))

plt.pie(severity_counts, labels=['Slight', 'Serious', 'Fatal'], autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'},textprops={'fontsize': 12}
    ,pctdistance=0.85 )

centre_circle = plt.Circle((0, 0), 0.50, fc='white')
plt.gca().add_artist(centre_circle)

plt.title('Accident Severity Distribution', fontsize=20, fontweight='bold', color='black')

plt.show()
```

# Time Series Analysis

Trend Analysis Over Time

```python
plt.figure(figsize=(12, 6))
df.groupby('Date').size().plot(kind='line', color='crimson')
plt.title('Accidents Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

Accidents by month analysis

```python
monthly_accidents = df.groupby('Month')['Accident_Index'].count()
monthly_accidents
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
monthly_accidents.plot(kind='bar', color='#A7C7E7')
plt.title('Accidents Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=0)
plt.show()
```

Weekday trends Analysis

```python
df['Weekday'] = df['Date'].dt.day_name()
weekday_accidents = df.groupby('Weekday')['Accident_Index'].count()
weekday_accidents
```

```python
plt.figure(figsize=(10, 5))
weekday_accidents = weekday_accidents.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
weekday_accidents.plot(kind='bar', color='#A8E6CF')
plt.title('Accidents Per Weekday')
plt.xlabel('Weekday')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

Identifying high risk hours

```python
hourly_accidents = df.groupby('Hour')['Accident_Index'].count()
hourly_accidents.sort_values()
```

```python
plt.figure(figsize=(12, 5))
hourly_accidents.plot(kind='line', marker='o', color='green')
plt.title('Accidents Per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.grid()
plt.xticks(range(0, 24))
plt.show()
```

Severity of hours per weekday

```python
heatmap_data = df.pivot_table(index="Weekday", columns="Hour", values="Accident_Index", aggfunc="count")

week_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
heatmap_data = heatmap_data.reindex(week_order)

plt.figure(figsize=(12, 3))  # Bigger figure
sns.heatmap(heatmap_data, cmap= "flare", linewidths=0.02, annot=False, fmt=".0f")

plt.title("Traffic Accidents Heatmap (Hour vs. Weekday)", fontsize=16, fontweight="bold")
plt.xlabel("Hour of the Day", fontsize=13)
plt.ylabel("Day of the Week", fontsize=13)
plt.xticks(rotation=45, fontsize=11)
plt.yticks(rotation=0, fontsize=11)

plt.show()
```

Rush hour and Late hours difference

```python
rush_hours = df[(df['Hour'] >= 7) & (df['Hour'] <= 10) | (df['Hour'] >= 16) & (df['Hour'] <= 20)]
late_night = df[(df['Hour'] >= 22) | (df['Hour'] <= 4)]


rush_count = rush_hours.shape[0]
late_count = late_night.shape[0]

plt.figure(figsize=(6, 4))
plt.bar(["Rush Hours", "Late Nights"], [rush_count, late_count], color=['#A7C7E7', '#A8E6CF'])
plt.xlabel("Time Period")
plt.ylabel("Number of Accidents")
plt.title("Accidents: Rush Hours vs Late Nights")
plt.show()
```

Location based Analysis

```python
# Group by Latitude & Longitude
top_locations = df.groupby(['Latitude', 'Longitude']).size().reset_index(name='Accident_Count')

top_10 = top_locations.nlargest(10, 'Accident_Count')
print(top_10)
```

Finding top dangerous locations as per road number

```python
top_locations = df['1st_Road_Number'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_locations.index, y=top_locations.values, palette="magma")
plt.xlabel("Road Number")
plt.ylabel("Accident Count")
plt.title("Top 10 Most Dangerous Roads")
plt.show()
```
